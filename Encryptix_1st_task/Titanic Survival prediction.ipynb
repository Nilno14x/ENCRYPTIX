```python
data1 = data.drop(columns=["PassengerId", "Cabin", "Name", "Ticket","AgeGroup"])
```

```python
data1["Age"] = data1["Age"].fillna(data["Age"].mean())
```

```python
data1['Fare'] = data1['Fare'].fillna(data['Fare'].mean())
```

```python
data1["Embarked"] = data1["Embarked"].fillna(method='backfill')
```

```python
data1.isnull().sum()
```

```python
data1.head()
```

```python
data1['Survived'] = data1['Survived'].map({0: 0, 1: 1})

data1['Pclass'] = data1['Pclass'].map({1: 1, 2: 2, 3: 3})

data1['Sex'] = data1['Sex'].map({'male': 1, 'female': 0})

data1['Embarked'] = data1['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

data1.tail()
```

## Splitting Independent and Dependent Variable

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay
```

```python
x = data1.drop(['Survived'],axis=1)
y = data1['Survived']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state= 20)
```

```python
x_train.head()
```

### LogisticRegression Model

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
lr_train = round(lr.score(x_train, y_train) * 100, 2)
print("Training Accuracy    :",lr_train ,"%")
```

```python
lr_accuracy = round(accuracy_score(y_pred_lr, y_test) * 100, 2)
print("Model Accuracy Score :",lr_accuracy ,"%")
```

```python
print("Classification_Report: \n",classification_report(y_test,y_pred_lr))
```

```python
cm = confusion_matrix(y_test, y_pred_lr)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=lr.classes_)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```

### KNeighborsClassifier Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred_knn = knn.predict(x_test)
```

```python
knn_train = round(knn.score(x_train, y_train) * 100, 2)
print("Model training Accuracy    :",knn_train ,"%")
```

```python
knn_accuracy = round(accuracy_score(y_pred_knn, y_test) * 100, 2)
print("Model Accuracy Score :",knn_accuracy ,"%")
```

```python
print("Classification_Report: \n",classification_report(y_test,y_pred_knn))
```

```python
cm = confusion_matrix(y_test, y_pred_knn)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot(cmap="Oranges")
plt.title("Confusion Matrix")
plt.show()
```

###  Support Vector Machines Model

```python
from sklearn.svm import SVC
```

```python
sv = SVC()
sv.fit(x_train, y_train)
```

```python
y_pred_sv = sv.predict(x_test)
```

```python
sv_train = round(sv.score(x_train, y_train) * 100, 2)
print("Training Accuracy    :",sv_train ,"%")
```

```python
sv_accuracy = round(accuracy_score(y_pred_sv, y_test) * 100, 2)
print("Model Accuracy Score :",sv_accuracy ,"%")
```

```python
print("Classification_Report: \n",classification_report(y_test,y_pred_sv))
```

```python
cm = confusion_matrix(y_test, y_pred_sv)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=sv.classes_)
disp.plot(cmap="Greens")
plt.title("Confusion Matrix")
plt.show()
```

### Comparing Different Models

```python
models = pd.DataFrame({
    'Model': [
        'Logistic Regression',
        'KNeighborsClassifier',  
        'Support Vector Machines'
    ],
    'Training Accuracy': [
        lr_train, knn_train, sv_train
    ],
    'Model Accuracy Score': [
        lr_accuracy, knn_accuracy,sv_accuracy
    ]
})
```

```python
pd.set_option('display.float_format', '{:0.2f}'.format)

styled_models = (
    models
    .sort_values(by='Model Accuracy Score', ascending=False)
    .style
    .background_gradient(cmap='coolwarm')
    .hide(axis="index")
    .set_properties(**{
        'font-family': 'Arial',
        'font-size': '15px',
        'text-align': 'left'
    })
)

styled_models
```

```python
models=models.sort_values(by='Model Accuracy Score',ascending=False)[:3]
plt.figure(figsize=(8, 4))

sb.barplot(y= 'Model', x= 'Model Accuracy Score', data= models, width=0.4)
plt.title('COMPARE THE MODEL')
plt.xlabel('MODEL')
plt.ylabel('Training Accuracy');
```
