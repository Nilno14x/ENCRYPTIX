```python
genres_count=data['Genre'].str.split(',',expand=True)
genres_count.head(5)
```

```python
from collections import Counter

flattened_genres = genres_count.stack().str.strip().dropna()

genre_counts = Counter(flattened_genres)

sorted_genre_counts = dict(sorted(genre_counts.items()))

for genre, count in sorted_genre_counts.items():
    print(f"{genre}: {count}")
```

```python
data['Votes'] = data['Votes'].str.replace(',', '').astype(int)
data["Votes"].head()
```

```python
data['Year'] = data['Year'].str.extract(r'\((\d{4})\)').astype(int)
```

```python
#data['Duration'] = data['Duration'].str.extract(r'\((\d{3})\)').fillna(0).astype(int)
```

```python
data.info()
```

```python
sb.set(style = "darkgrid", font = "Lucida Calligraphy")
```

```python
year_counts = data['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
ax = sb.lineplot(x=year_counts.index, y=year_counts.values)

ax.set_title("Annual Movie Release Counts Over Time")
ax.set_xlabel("Years")
ax.set_ylabel("Count")

positions = range(year_counts.index.min(), year_counts.index.max() + 1, 5)
ax.set_xticks(positions)
ax.set_xticklabels(positions, rotation=90)

plt.show()
```

```python
plt.figure(figsize=(10, 5))
sb.histplot(data['Year'], bins=range(min(data['Year']), max(data['Year']) + 1), kde=False)

plt.title("Distribution of Movie Releases by Year")
plt.xlabel("Years")
plt.ylabel("Count")

plt.xticks(rotation=90)
plt.show()
```

```python
bar = data.sort_values('Rating', ascending=False)

plt.figure(figsize=(12, 6)) 

plt.barh(bar['Name'].head(10), bar['Rating'].head(10), align='center', color='#148f77')
plt.gca().invert_yaxis()  
plt.xlabel("Rating", fontsize=14, family='Lucida Calligraphy')
plt.title("Top 10 Movies by Rating", fontsize=16,  family='Lucida Calligraphy')

plt.show()
```

```python
top_actors = actors.head(20)

plt.figure(figsize=(12, 6))

plt.barh(top_actors.index, top_actors.values, color='#9b59b6', edgecolor='black')
plt.gca().invert_yaxis()

plt.xlabel("Number of Appearances", fontsize=14, fontweight='bold')
plt.title("Top 20 Most Frequent Actors in Movies", fontsize=16, fontweight='bold')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
top_ten = data["Director"].value_counts().nlargest(10)
    
plt.figure(figsize=(12, 6))
top_ten.plot(kind="bar", color="#ec7063", edgecolor="black")
    
plt.xticks(rotation=45, ha="right")
plt.title(f"Top 10 Director", fontsize=16, weight="bold")
plt.xlabel("Director", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
sb.set_style('darkgrid') 

plt.figure(figsize=(12, 6), facecolor='whitesmoke') 
plt.scatter(data['Year'], data['Votes'], color='royalblue', alpha=0.8) 
plt.title("Votes vs Year", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Votes", fontsize=14)
plt.grid(True)  
plt.xticks(rotation=45) 
plt.tight_layout()  
plt.show()
```

```python
top_50_movies = data.nlargest(50, 'Votes')

sb.set_style('darkgrid')

plt.figure(figsize=(12, 8), facecolor='whitesmoke')

plt.barh(top_50_movies['Name'], top_50_movies['Votes'], color='royalblue')
plt.title("Top 50 Movies by Votes", fontsize=16)
plt.xlabel("Number of Votes", fontsize=14)
plt.ylabel("Movie Name", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7) 
plt.tight_layout() 
plt.show()
```

```python
sb.set(style='whitegrid')  
sb.pairplot(data, diag_kind='kde') 
plt.suptitle("Pairplot of DataFrame", y=1.02) 
plt.show()
```
